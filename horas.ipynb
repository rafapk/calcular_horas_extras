```python
def calcularhora (lista_hora):
    timeList = lista_hora
    mysum = datetime.timedelta()
    for i in timeList:
        (h, m, s) = i.split(':')
        d = datetime.timedelta(hours=int(h), minutes=int(m), seconds=int(s))
        mysum += d
    #print(str(mysum))

    # fazendo split na reposta mysum
    data = str(mysum)
    data_split= data.split()
    #print(data_split)


    # Verificas existencia de dias completos hora>24
    if (len(data_split)>1):

        # fazendo split na parte das horas de sobra  < day (24hrs)
        horas_sobra=(data_split[2]).split(':')
        #print(horas_sobra)





        #multiplicando a quantidade de dias * 24 e somando as horas de sobra
        soma=  (int(data_split[0])*24) + (int(horas_sobra[0]))
        #print(soma)
        output= str(soma)+':'+horas_sobra[1]+':'+ horas_sobra[2] #adicionando os minutos e segundos
        #print(output)
        return output

    # Se hora<24
    else:

        output=str(mysum)
        #print(output)
        return output
        
def verificar_unidade_tempo (coluna):
    
    # incluir segundos se eles nao existirem 
    for index, row in df_excel_ne.iterrows():

        horas_extras =row[coluna]
        extras= horas_extras.split(':')

        # Verificas se ja esta incluso segundos:
        if (len(extras)<3):
            horas_extras=str(horas_extras)+":00"
            df_excel_ne.loc[index,coluna]=horas_extras

        # Verificas se ja esta incluso minutos:
        elif(len(extras)<2):
            horas_extras=str(horas_extras)+":00"
            df_excel_ne.loc[index,coluna]=horas_extras
```
